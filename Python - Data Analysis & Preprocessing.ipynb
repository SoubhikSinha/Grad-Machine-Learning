5 ðŸ”½. Handle outliers. Detect and manage outliers within the dataset.

For example, in the penguins dataset, while flipper lengths typically fall
within the range of [180 â€“ 210], certain entries might exhibit values like
[10, 30]. These can be considered outliers. Possible solutions:</br>

â€¢ Remove outliers. If there are just a few outliers is limited, you may
eliminate the rows containing these outliers. </br>
â€¢ Impute outliers. Replace the outliers with the mean/median/mode of the
feature.

```python
# Please note : outliers for features that contain categorical values
# doesn't make any sense. Thus, we will stick to the features that
# contain numerical values

# Lets see what all features have numerical values
print(features_list)
```

```python
# We will first create box plots to see abnormal values

# NOTE : The value in "calorie requirement" column has string values. Thus, they
# need to be converted to decimal values

for i in range(0, pf_df_copy['calorie requirement'].count()):
  pf_df_copy['calorie requirement'][i] = float(pf_df_copy['calorie requirement'][i])
```

```python
pf_df_copy
```

```python
# Lets create boxplots

for i in features_list:
  print("\n\n\n"+i+" ðŸ”½")
  plt.boxplot(pf_df_copy[i])
  plt.show()
```

```python
# We need to remove all the outliers from all the features / columns
# Thus, we need a function for the same

def outliers_removal(feature):
    Q1 = np.percentile(feature, 25)
    Q3 = np.percentile(feature, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    cleaned_data = []
    # cleaned_data = [i for i in feature if lower_bound <= i <= upper_bound]
    for i in feature:
      if (lower_bound <= i <= upper_bound):
        cleaned_data.append(i)
      else:
        cleaned_data.append('NaN')

    return cleaned_data

# Now lets put this function into action !

for i in features_list:
  pf_df_copy[i] = outliers_removal(pf_df_copy[i])
```

```python
# Lets print the dataset

pf_df_copy
```

```python
# Now we will replace the 'NaN' values with the mean values of the features
# As it will be less skewed.

for i in features_list:
  for j in range(0, pf_df_copy[i].count()):
    if (pf_df_copy[i][j] == 'NaN'):
      pf_df_copy[i][j] = np.nan
```

```python
pf_df_copy
```

```python
features_mean = []

for i in features_list:
  features_mean.append(pf_df_copy[i].mean())

for i,j in zip(features_list, features_mean):
    pf_df_copy[i].fillna(value = j, inplace = True)
```

```python
# Creating the list of minimum & maximum values of all the existing features
# that contain numerical values

# We are excluding the features, which were removed from the
# dataset because of negative correlation

min_list = []
max_list = []
for i in features_list:
  min_list.append(diamond_df_copy[i].min())
  max_list.append(diamond_df_copy[i].max())
```

```python
# Printing the minimum & maximum values lists

print(min_list)
print(max_list)
```

```python
# Lets normalize the data

for i,k,l in zip(features_list, min_list, max_list):
  for j in range(0, 53940):
    diamond_df_copy[i][j] = ((diamond_df_copy[i][j] - k)/(l-k))
```
